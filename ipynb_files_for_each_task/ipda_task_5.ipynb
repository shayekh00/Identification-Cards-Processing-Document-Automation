```python
# Step2. import libraries
import pytesseract
import os
from PIL import Image

#for visualising
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step3. Upload Image to the Colab
dirname = os.path.dirname(__file__)
base_path = os.path.join(dirname, 'correct_path_to_dataset/5_ocr/5_ocr/') # define the correct path
fake_ids = sorted(os.listdir(base_path))

counter = 0
debug = True
for img_path in fake_ids:
  if counter >= 13:
    img_path = os.path.join(base_path,img_path)
    img = Image.open(img_path).convert('L')

    if debug:
      img_array = np.array(img)

      plt.figure(figsize=(8, 8))
      #plt.subplot(1, 3, 1)
      plt.imshow(img_array, cmap='gray')
      plt.title('Train Image')
      plt.axis('off')

    break
  counter = counter + 1
```

```python
#my_config = '--psm 5 -l deu'
#my_config = '--psm 13'
# standard psm (i.e. --psm 4) works the best
extractedInformation1 = pytesseract.image_to_string(img, lang='deu')
extractedInformation2 = pytesseract.image_to_string(img)
print("German Language:")
print("--------------------")
print(extractedInformation1)
print("--------------------\n")
print("English Language:")
print("--------------------")
print(extractedInformation2)
print("--------------------")
```
